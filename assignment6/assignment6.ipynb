```python
def dxdt(x,v,t):
    return v

def dvdt(x,v,t,l=0.2,w=1.2):  # differential equation
    import math
    return (2*math.cos(w*t)-2*x-l*v)/2

X, T, i = ml.coup_ode2(dxdt,dvdt,2,-1,0,0.1,120)
plt.scatter(T, X,s=2)
#plt.plot(T,X)
plt.xlabel('T')
plt.ylabel('y(T)')
```

# 2

```python
def dydx(y,z,x):
    return z

def dzdx(y,z,x):  #diff eqn
    return -0.01*(20-y)

X = np.linspace(0, 10, 100) # point generations
y1 = []
for i in range(len(X)):
    a = ml.bvp(dydx, dzdx, X[i],200,50,40,200) 
    y1.append(a)

e=0.3 #precision
for i in range(len(y1)):
    if 100-e<y1[i]<100+e:
        print('Length for which temperature is 100+/-',e,'=',round(X[i],3))
        
plt.scatter(X, y1, marker='+')
plt.ylabel('Temperature $(^\circ C)$')
plt.xlabel('Length')
plt.show()
```
