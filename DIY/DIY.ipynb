```python
import MyLibrary as ml
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x): #function
    return math.exp(-x)
```

```python
ylist=[]
xlist=[]
ilist=np.linspace(0,10,10)
for i in range(1,10):
    A=ml.adaptive_simpson(f,0,4,10**(-i),0)
    ylist.append(A[0])#integral
    xlist.append(A[1])#iter for that precision
    
plt.figure(figsize=(7,5))
plt.scatter(xlist,ylist)
plt.plot(xlist,ylist)
plt.xlabel('No. of iterations for precisions from $10^{i}$ , i$\in$[0,10]')
plt.ylabel('Integral')
plt.title("Convergence for adaptive Simpson's method")
plt.savefig('asimpson',dpi=1000)
```

```python
p=[]
q=[]
for i in range(10):
    N=ml.nsimpson(0,4,10**(-i),1) #fdermax=1 for exp(-x)
    s1=ml.simpson(f,0,4,N)
    p.append(s1) # list p contains integral
    q.append(N) # no. of iterations

plt.figure(figsize=(7,5))
plt.scatter(q,p)
plt.plot(q,p)
plt.xlabel('No. of iterations for precisions from $10^{i}$ , i$\in$[0,10]')
plt.ylabel('Integral')
plt.title("Convergence for Simpson's method")
plt.savefig('simpson',dpi=1000)
```

```python
plt.figure(figsize=(7,5))
plt.plot(q,p,label="Simpson's") # Zoomed view of above plots
plt.scatter(q,p)
plt.xlim(0,50)
plt.ylim(0.976,0.984)
plt.scatter(xlist,ylist)
plt.plot(xlist,ylist,label="Adaptive Simpson's")
plt.legend()
plt.xlabel('No. of Iterations')
plt.ylabel('Integral')
plt.savefig('zoomed.jpg',dpi=1000)
```
