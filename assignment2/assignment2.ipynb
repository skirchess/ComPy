```python
import MyLibrary as ml
import matplotlib.pyplot as plt
with open('input_3.txt') as f: # input_3.txt contains input for question 1,2,3,5
    lines = [ line.strip() for line in f ]
    list1=[]
    for k in lines:
        list1.append(float(k))
```

# 1

```python
randnum=ml.lcg(list1[0],list1[1],list1[2],list1[3],list1[7])
print(randnum)
```

# 2

```python
x=ml.lcg(list1[0],list1[1],list1[2],list1[3],list1[6]) #list1[3],list1[4],list1[5] are seeds
y=ml.lcg(list1[0],list1[1],list1[2],list1[4],list1[6])
z=ml.lcg(list1[0],list1[1],list1[2],list1[5],list1[6])

#Throwing Method
counter=0
for i in range(len(x)): #divding random numbers by m
    x[i],y[i],z[i]=x[i]/list1[2],y[i]/list1[2],z[i]/list1[2]
    
    if (x[i])**2+(y[i])**2+(z[i])**2<=1:
        counter=counter+1
pi=6*(counter/(len(x))) #len(x) is total trails
print('π =',pi)
print('volume of a octant in unit sphere=',pi/(6))
```
